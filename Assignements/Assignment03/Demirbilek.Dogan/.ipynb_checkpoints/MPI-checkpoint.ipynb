```python
import numpy as np
from mpi4py import MPI

comm = MPI.COMM_WORLD
rank = comm.Get_rank()
size = comm.Get_size()
```

```python
xR = +2
xL = -2
yR = +2
yL = -2
nx = 100
ny = 100
```

```python
dx = (xR - xL) / nx
dx
```

```python
dy = dy = (yR - yL) / ny
dy
```

```python
rank = [0,1,2,3]
size = 4
```

```python
for i in rank:
    ranstarty = i * (ny // size)  
    ranstartx = i * (nx // size)
    
    ranendy = (i + 1) * (ny // size) + (ny % size > i)  
    ranendx = (i + 1) * (nx // size) + (nx % size > i)
    
    print(ranstarty, ranendy)
    #print("\t",ranstartx, ranendx)
        
    for y in range(ranstarty,ranendy):
        for x in range(ranstartx,ranendx):
            c = (xL + (x*dx),yL + (y*dy))
            print(i,">>>>>>>>",c)
```
